## Web Scrapping Script for Ratings Prediction Project 

```python
                                           ''' Importing required libraries'''
    
import selenium
from bs4 import BeautifulSoup
import time
import pandas as pd


# Importing selenium webdriver
from selenium import webdriver

# Importing Exceptions
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException, ElementClickInterceptedException, ElementNotVisibleException

# importing requests
import requests

#importing regex
import re

import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)


from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

```python
#connecting to the webdriver
driver = webdriver.Chrome(r"C:\Users\Rohit\Documents\chrome_webdriver\chromedriver.exe")

#Website
amazon='https://www.amazon.in/'

#creating empty lists to store the data
amzn_urls = []
ratings_amzn = []
rate_urls = []
reviews_amzn = []
rev_urls = []

#getting the amazon webpage
driver.get(amazon)
time.sleep(2)
    
#creating list of products
products = ['Laptops', 'Smartphones', 'Headphones', 'Smartwatches']

for p in products:
    search_bar=driver.find_element_by_xpath('//input[@id="twotabsearchtextbox"]')
    search_bar.send_keys(p)
    search_btn=driver.find_element_by_xpath('//input[@id="nav-search-submit-button"]')
    search_btn.click()
    time.sleep(2)

    #getting url tags 
    url_tags = driver.find_elements_by_xpath('//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    #appending urls into the list
    for u in url_tags:
        try: 
            amzn_urls.append(u.get_attribute("href"))
        except:
            amzn_urls.append('-')
        
    #getting the ratings
    for i in amzn_urls:
        driver.get(i)
        time.sleep(2)
        #getting the url tags for ratings pages
        try:
            url_tag5 = driver.find_element_by_xpath("//a[@class='a-link-normal 5star']")
            url_tag4 = driver.find_element_by_xpath("//a[@class='a-link-normal 4star']")
            url_tag3 = driver.find_element_by_xpath("//a[@class='a-link-normal 3star']")
            url_tag2 = driver.find_element_by_xpath("//a[@class='a-link-normal 2star']")
            url_tag1 = driver.find_element_by_xpath("//a[@class='a-link-normal 1star']")
            #appending urls into the list
            rate_urls.append(url_tag5.get_attribute("href"))
            rate_urls.append(url_tag4.get_attribute("href"))
            rate_urls.append(url_tag3.get_attribute("href"))
            rate_urls.append(url_tag2.get_attribute("href"))
            rate_urls.append(url_tag1.get_attribute("href"))
        except NoSuchElementException:
                        pass
    
    
    #getting the tags for reviews
    for f in rate_urls:
        driver.get(f)
        try:
            for j in range(0,6):
                time.sleep(2)
                revw_tag = driver.find_elements_by_xpath("//span[@class='a-size-base review-text review-text-content']")
                rate_tag = driver.find_elements_by_xpath("//a[contains(@href,'/gp/customer-reviews/')]/i/span[1]")
        except NoSuchElementException or InvalidArgumentException:
                pass
        for k in revw_tag:
            try:
                reviews_amzn.append(k.text.replace("\n"," "))
            except:
                reviews_amzn.append("-")
        for r in rate_tag:
            try:
                ratings_amzn.append(r.get_attribute("innerHTML"))
            except:
                ratings_amzn.append("-")
        try:
            nxtbtn = driver.find_element_by_xpath("//*[@id='cm_cr-pagination_bar']/ul/li[2]/a").click()
            driver.refresh()
        except NoSuchElementException:
                pass
        
        
    #filtering only ratings. Removing text from ratings
    Ratings_amzn=[]
    for l in ratings_amzn:
        Ratings_amzn.append(l.replace(' out of 5 stars',''))
        time.sleep(1)
    
    #getting the amazon webpage
    driver.get(amazon)
    time.sleep(2)
    driver.refresh()
```
